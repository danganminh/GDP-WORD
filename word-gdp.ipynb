```python
data = data.dropna().reset_index(drop=True)
```

```python
def return_Mean_Economic_Growth_Rate(df):
    name = []
    values = []
    for i in range(len(df)):
        name.append(data['GDP, current prices (Billions of U.S. dollars)'][i])
        GDP = data.drop(labels=['GDP, current prices (Billions of U.S. dollars)'], axis=1).iloc[i].to_list()
        EGR = []
        for j in range(len(GDP) - 1):
            try:
                val = (GDP[j+1] / GDP[j]) * 100 - 100
            except:
                val = 0
            EGR.append(val)
        values.append(np.mean(EGR))
    
    return pd.DataFrame({
        'Country': name,
        'Mean_Economic_Growth_Rate': values
    })
```

```python
Mean_Economic_Growth_Rate = return_Mean_Economic_Growth_Rate(data)
display(Mean_Economic_Growth_Rate)
```

```python
Mean_Economic_Growth_Rate = Mean_Economic_Growth_Rate.sort_values(by=['Mean_Economic_Growth_Rate'], ascending=False).reset_index(drop=True)
Mean_Economic_Growth_Rate
```

# Autoregression

```python
from statsmodels.tsa.ar_model import AutoReg
from sklearn.metrics import mean_squared_error
```

```python
columns = data.T.values[0]
gdp_values = data.T.values[1:]
data_T = pd.DataFrame(data=gdp_values, columns=columns)

data_T.head()
```

```python
def predict_auto(df, country, numyear_pre, test_size=0.2):
    
    gdp_values = df[country].to_list()
    len_test = int(len(gdp_values)*test_size)
    
    train_data = gdp_values[:len(gdp_values)-len_test]
    test_data = gdp_values[len(gdp_values)-len_test:]

    # Search best Lag(s)
    MSE_ar = []
    for i in range(1, 6):
        mod = AutoReg(train_data, lags=i)
        res = mod.fit()
        predictions = res.predict(start=len(train_data), end=len(train_data) + len(test_data) - 1)
        mse = mean_squared_error(test_data, predictions)
        MSE_ar.append(mse)
    best_lag = np.where(MSE_ar == np.min(MSE_ar))[0] + 1
    print('Best Lag(s):', best_lag)
    
    # Predict
    mod = AutoReg(gdp_values, lags=best_lag)
    res = mod.fit()
    predictions = res.predict(start=len(gdp_values), end=len(gdp_values) + numyear_pre - 1)

    return predictions, np.min(MSE_ar)
```

```python
country = 'United States'
numyear = 10
future, mse = predict_auto(data_T, country, numyear_pre=numyear)
rmse = np.sqrt(mse)

plt.figure(figsize=(15, 5))

# Historical data
x1 = np.arange(1980, 1980 + len(data_T[country]))
y1 = data_T[country].to_list()
plt.plot(x1, y1, color='blue', label='Historical Data')

# Future predictions
x2 = np.arange(1980 + len(data_T[country])-1, 1980 + len(data_T[country]) + len(future)-1)
y2 = future

plt.plot(x2, y2, color='red', label='Predictions')
plt.fill_between(x2, y2-rmse, y2+rmse, color='green', alpha=0.3, label='RMSE')

plt.xlabel('Year')
plt.ylabel('GDP')
plt.title(f'GDP of {country} Prediction {numyear} Years')
plt.legend()
plt.grid(True)
plt.show()
```

# Neural Networks

```python
import tensorflow as tf
```

```python
def model():
    model = tf.keras.Sequential([
        tf.keras.Input(shape=(1,)),
        tf.keras.layers.Dense(16, activation='linear'),
        tf.keras.layers.Dense(16, activation='linear'),
        
        tf.keras.layers.Dense(1)
    ])

    model.compile(optimizer=tf.keras.optimizers.Adam(),
                loss="mse",
                metrics=['mse'])
    
    return model
```

```python
gdp_values = data_T['United States'].dropna().astype(float)
len_test = int(len(gdp_values) * 0.2)

train_data = gdp_values[:len(gdp_values) - len_test].values
x_train = np.arange(1980, 1980 + len(train_data))

test_data = gdp_values[len(gdp_values) - len_test:].values
x_test = np.arange(1980 + len(train_data), 1980 + len(train_data) + len(test_data))

# Reshape the data for the model
x_train = x_train.reshape(-1, 1)
train_data = train_data.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
test_data = test_data.reshape(-1, 1)

model = model()

history = model.fit(x_train, train_data, epochs=500, verbose=2)

# Evaluate the model on test data
test_loss, test_mse = model.evaluate(x_test, test_data, verbose=2)
```

```python
test_loss, test_mse = model.evaluate(x_test, test_data, verbose=2)
print(f'Test MSE: {test_mse}')
```

```python
# Predict future GDP values
numyear_pre = 10
x_future = np.arange(1980 + len(gdp_values), 1980 + len(gdp_values) + numyear_pre).reshape(-1, 1)
future_gdp = model.predict(x_future)

# Plotting the results
plt.figure(figsize=(15, 5))

# Historical data
x1 = np.arange(1980, 1980 + len(gdp_values))
y1 = gdp_values.values
plt.plot(x1, y1, color='blue', label='Historical Data')

# Future predictions
x2 = np.arange(1980 + len(gdp_values), 1980 + len(gdp_values) + len(future_gdp))
y2 = future_gdp.flatten()
plt.plot(x2, y2, color='red', label='Predictions')

plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
